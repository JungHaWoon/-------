# 종합코드1
> * 해제시각이 있는 특보정보 만들기

```python
import pandas as pd
df = pd.read_csv('./first_Lv0_point 로 끝나는 csv _ 해양 기상특보가 시작되기 직전, 직후 시점까지 잡아낸 csv _ 2011_2022/W_REPORT_2012_ver07_02__first_Lv0_point.csv')

df1 = df[df['특보강도']==0]
df1.rename(columns={'일시':'해제시각'},inplace=True)
df1.drop(columns='특보강도',inplace=True)
df_concat = pd.concat([df,df1], axis=1)
df_concat['해제시각'].fillna(method='bfill', inplace=True)

df_concat.to_csv('./target_label2/2012.csv')
```

```python
import pandas as pd
df = pd.read_csv('./first_Lv0_point 로 끝나는 csv _ 해양 기상특보가 시작되기 직전, 직후 시점까지 잡아낸 csv _ 2011_2022/W_REPORT_2012_ver07_02__first_Lv0_point.csv')
empty_df = pd.DataFrame()
for i in range(len(df)-1) :
    if (df.iloc[i,1]!=df.iloc[i+1,1]) :
        empty_df = pd.concat([empty_df,df.iloc[[i+1]]])

empty_df.rename(columns={'일시':'해제시각'},inplace=True)
empty_df.drop(columns='특보강도',inplace=True)
df_concat = pd.concat([df,empty_df], axis=1)
df_concat['해제시각'] = df_concat['해제시각'].shift(-1)
df_concat['해제시각'].fillna(method='bfill', inplace=True)

df_concat.to_csv('./target_label2/2012.csv')
```

# 작업전체 종합코드

```python
import pandas as pd

df1 = pd.read_csv('./등표기상관측_순서정리/2011.csv') 
df2 = pd.read_csv('./target_label2/2011.csv',index_col=0)
df1['일시'] = pd.to_datetime(df1['일시'])
df2['일시'] = pd.to_datetime(df2['일시'])

df = pd.merge(df1,df2,how='outer',on='일시')
df.sort_values('일시',inplace=True)
df.loc[0,'특보강도'] = 0
df['특보강도'].fillna(method='ffill',inplace=True)
df['해제시각'].fillna(method='ffill',inplace=True)
df.drop(df[df['특보강도']==0].index,inplace=True)
df.dropna(subset=['WIND_SPEED'],inplace=True)

df['일시'] = pd.to_datetime(df['일시'])
df['해제시각'] = pd.to_datetime(df['해제시각'])
df['해제예상'] = df['해제시각'] - df['일시']

df.set_index('일시',inplace=True)
df.drop(columns='해제시각',inplace=True)
df.to_csv('./모델2_데이터셋/2011.csv')
```

```python
import pandas as pd
df = pd.read_csv('./모델2_데이터셋./2011.csv', index_col=0)

df['day'] = df['해제예상'].str[0:1]
df['hour'] = df['해제예상'].str[7:9]
df['minute'] = df['해제예상'].str[10:12]

df['day'] = df['day'].astype('float')
df['hour'] = df['hour'].astype('float')
df['minute'] = df['minute'].astype('float')

df['day'] = df['day']*24
df['minute'] = df['minute']/60

df['남은시간'] = df['day']+df['hour']+df['minute']

df.drop(columns=['해제예상','day','hour','minute'],inplace=True)

df.to_csv('./모델2_데이터셋ver2/2011.csv')
```
